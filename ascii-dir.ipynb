```python
import os

def generate_ascii_tree(path, level=0, prefix=''):
    # Sort the contents of the directory to maintain the order of the items.
    items = sorted(os.listdir(path))

    for i, item in enumerate(items):
        full_path = os.path.join(path, item)

        # Ignore .ipynb_checkpoints folders
        if item == '.ipynb_checkpoints':
            continue

        # If the current item is a directory, display it with a '|' and '--'
        if os.path.isdir(full_path):
            if i == len(items) - 1:
                print(f'{prefix}└── {item}/')
                new_prefix = f'{prefix}    '
            else:
                print(f'{prefix}├── {item}/')
                new_prefix = f'{prefix}│   '

            # Recursively call the function for the directory contents.
            generate_ascii_tree(full_path, level + 1, new_prefix)

        # If the current item is a file, display it with a '|' and '--'
        elif level != 0:
            if i == len(items) - 1:
                print(f'{prefix}└── {item}')
            else:
                print(f'{prefix}├── {item}')

if __name__ == '__main__':
    root_path = '.'  # The root directory of your repository
    generate_ascii_tree(root_path)

```
